### Print the first 10 rows of the data. 

```python
cars = pd.read_csv("cars_data.csv")

print(cars.head(10))
```

### Print the last 5 rows. 

```python
print(cars.tail())
```

### By using the info method, check how many non-null rows each column have. 

```python
cars.info()
```

### If any column has a missing value, drop the entire row. Notice, the operation should be inplace meaning you change the dataframe itself.

```python
# rem_all = cars.dropna(how='all')    #Dropping rows if all values in that row are missing.

# Drop rows with at least one Nan value (Not a Number value). Using dropna() function and assigning to new var
rem = cars.dropna()                 
rem.info()
```

```python
# Same but inplace
cars.dropna(inplace=True)
cars.info()
```

### Calculate the mean of each numeric column. 

```python
# print(cars.describe())
print("\nJust the mean values:")
cars.mean(numeric_only=True)
```

### Select the rows where the column "company" is equal to 'honda'. 

```python
#Using Masking to find rows with honda
cars[cars.loc[:,'company'] =='honda']
```

### Sort the data set by price in descending order. This should *not* be an inplace operation. 

```python
cars.sort_values(by='price', ascending=False)
```

### Select the rows where the column "company" is equal to any of the values (audi, bmw, porsche).

```python
# Select Rows where Column Value is in List of Values
# Using masking with function 'isin'
cars[cars.loc[:,'company'].isin(['audi','bmw','porsche'])]
```

### Find the number of cars (rows) for each company. 

```python
#Longer version
cars.groupby('company').agg('index').count()

#Shorter version
cars.value_counts('company')
```

### Find the maximum price for each company. 

```python
cars.groupby('company').agg('price').max()
```
