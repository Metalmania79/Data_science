### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

```python
'''
    Explicitly create Figures and Axes, and call methods on them (the "object-oriented (OO) style").
    Rely on pyplot to implicitly create and manage the Figures and Axes, and use pyplot functions for plotting.

    Recommended approach is to use Object Oriented style
'''
```

### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

```python
'''
Figure - Where the data is graphed
Axes - An area where points can be specified in terms of for example X-Y/X-Y-Z coordinates
Axis - One of many directions, X-axis, Y-axis, Z-axis etc.
Artists - Another term for objects being created on the canvas/figure. Text labels, line objects etc.
'''
```

### When plotting in Matplotlib, what is the expected input data type?

```python
'''
Numpy arrays or objects that can be passed to numpy.asarray. 
Pandas, data objects and numpy.matrix may not work as intended. They should be converted to numpy.array/numpy.asarray
'''
```

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
x = np.linspace(-2,2,10)
y = x**2

## Object Oriented approach
figure, ax = plt.subplots()
ax.plot(x, y, label='En linje')
ax.plot(x, y/2, label='En linje till')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.legend()

##Pyplot approach
plt.figure(figsize=(5,3), layout='constrained')
plt.plot(x, y, label='Blå linje')
plt.plot(x, y/2, label='Orange linje')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())
```

```python
figure, ax = plt.subplots(1,2, figsize=(8,4), layout='constrained')
bars = ax[0].bar(names, values, label='Amount of fruit', color='red',edgecolor='gray')
scatter = ax[1].scatter(random_data_x, random_data_y, label='Green peas', facecolor='lightgreen', edgecolor='darkgreen')

#Creates a separate legend for bars and adds legend manually to the axes
first_legend = ax[0].legend(handles=[bars], loc='upper left')
ax[0].add_artist(first_legend)  

ax[0].set_xlabel('Fruits')
ax[0].set_ylabel('Amount')
ax[0].set_title('Fruits and greens')
ax[0].annotate('Look grapes!', xy=(0, 20), xytext=(1, 20),arrowprops=dict(facecolor='Yellow', shrink=0.15))
ax[1].legend(handles=[scatter], loc='lower left')

# Different approach and not separated
plt.figure(figsize=(5,3), layout='constrained')
plt.bar(names, values, label='Fruity bars', color='Orange',edgecolor='gray')
plt.scatter(random_data_x, random_data_y, label='Green balls',facecolor='lightgreen', edgecolor='darkgreen')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('A big mess of everything')
plt.legend()
```
